### We will study a political blogs dataset first compiled for the paper Lada A. Adamic and Natalie Glance, “The political blogosphere and the 2004 US Election”, in Proceedings of the WWW-2005 Workshop on the Weblogging Ecosystem (2005). The dataset nodes.txt contains a graph with n = 1490 vertices (“nodes”) corresponding to political blogs. Each vertex has a 0-1 label (in the 3rd column) corresponding to the political orientation of that blog. We will consider this as the true label and try to reconstruct the true label from the graph using the spectral clustering on the graph. The dataset contains edges between the vertices. You may remove isolated nodes (nodes that are not connected any other nodes). ###

### Using spectral clustering to find the $k = 2$, $k = 3$, and $k = 4$ clusters in the network of political blogs (each node is a blog, and their edges are defined in the file edges.txt). We will treat the network as an undirectly graph; thus, when constructing the adjacency matrix, make sure to it is symmetrical. Then report the majority labels in each cluster, when $k = 2, 3, 4$, respectively. For example, if there are $k = 2$ clusters, and their labels are $\{0, 1, 1, 1\}$ and $\{0, 0, 1\}$ then the majority label for the first cluster is 1 and for the second cluster is 0. ###


### The first step is build the symmetric adjacency matrix (A) based on the edges information from the dataset. Note that there are 266 isolated nodes. After removing those isolated nodes, (the corresponding columns and rows from the A matrix) we have the updated A matrix of size 1224 × 1224. We can then calculate the Laplacian matrix L = D − A accordingly. Next, for k = 2, 3, 4, respectively, we use k eigenvectors for the smallest k eigenvalues of L, perform k-means and report the majority labels in each cluster. ###


- (a) k = 2. Cluster 1: the majority label is 1; Cluster 2: the majority label is 0.
- (b) k = 3. Cluster 1: the majority label is 1; Cluster 2: the majority label is 1; Cluster 3: the
majority label is 0.
- (c) k = 4. Cluster 1: the majority label is 1; Cluster 2: majority is 1; Cluster 3: majority is 1; Cluster
4: majority is 0.





### Let's compare the majority label with the individual labels in each cluster, and evaluate the mismatch rate for each cluster, when $k = 2, 3, 4$. For instance, in the example above, the mismatch rate for the first cluster is 1/4 (only the first node differs from the majority) and the the second cluster is 1/3 ###

- (a) k = 2. Cluster 1: the mismatch rate is 0.4795; Cluster 2: the mismatch rate is 0.
- (b) k = 3. Cluster 1: the mismatch rate is 0.4811; Cluster 2: the mismatch rate is 0; Cluster 3: the
mismatch rate is 0.
- (c) k = 4. Cluster 1: the mismatch rate is 0.4819; Cluster 2: the mismatch rate is 0; Cluster 3: the
mismatch rate is 0; Cluster 4: the mismatch rate is 0.
